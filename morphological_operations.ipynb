```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('images/j.png', cv.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
```

```python
# Make the kernel
kernel = np.ones((5, 5), np.uint8)
```

```python
# Erosion
erosion = cv.erode(img, kernel, iterations=1)
plt.imshow(erosion, cmap='gray')
```

```python
# Dilation
dilation = cv.dilate(img, kernel, iterations=1)
plt.imshow(dilation, cmap='gray')
```

```python
# Opening
img = cv.imread('images/j_for_opening.png', cv.IMREAD_GRAYSCALE)
opening = cv.morphologyEx(img, cv.MORPH_OPEN, kernel)
plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(opening,cmap = 'gray')
plt.title('Opened Image'), plt.xticks([]), plt.yticks([])
```

```python
# Closing
img = cv.imread('images/j_for_closing.png', cv.IMREAD_GRAYSCALE)
closing = cv.morphologyEx(img, cv.MORPH_CLOSE, kernel)
plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(closing,cmap = 'gray')
plt.title('Closed Image'), plt.xticks([]), plt.yticks([])
```

```python
# Morphological Gradient
# It is the difference between dilation and erosion of an image.
img = cv.imread('images/j.png', cv.IMREAD_GRAYSCALE)
gradient = cv.morphologyEx(img, cv.MORPH_GRADIENT, kernel)
plt.imshow(gradient, cmap='gray')
```
