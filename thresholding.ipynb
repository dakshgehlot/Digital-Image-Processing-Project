```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
# cv.IMREAD_GRAYSCALE: Loads image in grayscale mode
img = cv.imread("images/cameraman.tif", cv.IMREAD_GRAYSCALE)
assert img is not None, "Could not read the image"
```

```python
# Simple thresholding. Returns two values: the threshold used and the thresholded image
ret, th_img_1 = cv.threshold(img, 127, 255, cv.THRESH_BINARY)

# Adaptive thresholding, first using the mean of the neighborhood area and then using
# the gaussian weighted sum
th_img_2 = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 11, 2)
th_img_3 = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 11, 2)
```

```python
titles = ["Original Image", "Global Thresholding (v = 127)", "Adaptive Mean Thresholding", "Adaptive Gaussian Thresholding"]
images = [img, th_img_1, th_img_2, th_img_3]
```

```python
for i in range(4):
    plt.subplot(2, 2, i + 1), plt.imshow(images[i], "gray")
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
```
