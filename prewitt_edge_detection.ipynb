```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('images/messi.png', cv.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
```

```python
img_blurred = cv.GaussianBlur(img, (3, 3), 0)
```

```python
kernelx = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
img_prewittx = cv.filter2D(img_blurred, -1, kernelx)
img_prewitty = cv.filter2D(img_blurred, -1, kernely)
```

```python
plt.figure(figsize=(5, 5))
plt.subplot(2, 2, 1), plt.imshow(img, cmap='gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2, 2, 2), plt.imshow(img_blurred, cmap='gray')
plt.title('Blurred'), plt.xticks([]), plt.yticks([])
plt.subplot(2, 2, 3), plt.imshow(img_prewittx, cmap='gray')
plt.title('Prewitt X'), plt.xticks([]), plt.yticks([])
plt.subplot(2, 2, 4), plt.imshow(img_prewitty, cmap='gray')
plt.title('Prewitt Y'), plt.xticks([]), plt.yticks([])
plt.show()
```
