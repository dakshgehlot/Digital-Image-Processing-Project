```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/cameraman.tif', 0)
```

```python
imf = np.float32(img)
```

```python
# Find dct 
dct = cv2.dct(imf)
```

```python
# Apply inverse dct
idct = cv2.idct(dct)

# Convert to uint8
idct = np.uint8(idct)
```

```python
# Plot the images
plt.figure()
plt.subplot(131)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(132)
plt.imshow(dct, cmap='gray')
plt.title('DCT')
plt.axis('off')

plt.subplot(133)
plt.imshow(idct, cmap='gray')
plt.title('IDCT')
plt.axis('off')

plt.show()
```
